```python
import pandas as pd
df = pd.read_csv("/content/train.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop('Cabin',axis=1)
df
```

```python
df=df.dropna()
df
```

```python
df.shape
```

```python
Sex=pd.get_dummies(df['Sex'])
Sex
```

```python
Embarked=pd.get_dummies(df['Embarked'])
Embarked
```

```python
X=pd.concat([df,Sex, Embarked],axis=1)
X
```

```python
X=X.drop(['Sex','Embarked'],axis=1)
X
```

```python
import matplotlib.pyplot as plt
plt.bar(df['PassengerId'],df['Survived'],color='red')
```

```python
plt.bar(df['Name'],df['Age'])
```

```python
X=X.iloc[:,[0,2,4,5,6,9,10]].values
X
```

```python
Y=df.iloc[:,2]. values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0)
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

```python
Y_test
```

```python
df1=pd.DataFrame({'Actual':Y_test,'Predicted':Y_pred})
df1
```

```python
df1.plot(figsize=(15,8))
```

```python
df1.plot(figsize=(15,8),kind='bar')
```

```python
import seaborn as sns
sns.regplot(x='Actual',y='Predicted',data=df1,color='red')
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,Y_pred)
```
